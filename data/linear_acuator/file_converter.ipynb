```python
import scipy.io
import numpy as np
import pandas as pd
import os, re


MAT_FILE_PATH = ""


# Load the .mat file
for idx in range(1,3):
    b_name = "Backlash"
    full_file_path = MAT_FILE_PATH + b_name + str(idx) + '.mat'
    print(full_file_path)
    mat = scipy.io.loadmat(full_file_path)

    # Extract the data from the variable (replace 'your_variable_name' with the actual variable name)

    motion_profile = ["sin", "trap"]
    degradation_stage = range(1,9)
    load  =  ["20", "40", "neg40"]
    weigth_label = "kg"
    index = range(1,11)
    dataframes = []
    for motion in motion_profile:
        for l in load:
            for i in index:
                for deg in degradation_stage:
                    for step in ["st", "nd","rd", "th"]:
                        df = None
                        try:
                            colname = "back"+ motion + str(deg) + step + str(l) + weigth_label + str(i)
                            df = mat[colname]
                        except:
                            #print(colname)
                            continue

                        df = pd.DataFrame(df)

                        
                        df['profile'] = motion
                        if l == "neg40":
                            df['load'] = -40
                        else:
                            df['load'] = int(l)
                        df["run"] = i
                        base_name = os.path.basename(full_file_path)
                        numbers = re.findall(r'\d+\.\d+|\d+', base_name)
                        df["fault_type"] = numbers[-1]
                        dataframes.append(df)

    # Save as CSV
    #dataframes
    new_data = pd.concat(dataframes, ignore_index=True) 
    new_data = new_data.rename(columns={0: 'set_point ', 1: 'position_error', 2: 'current'})
    new_data.to_csv('raw' + b_name +str(idx)  +  '.csv', index=False)
```

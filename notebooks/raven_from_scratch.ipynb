```python
import numpy as np
import matplotlib.pyplot as plt
from typing import Tuple
```

```python
# Function to generate a Morlet wavelet
def morlet_wavelet(t, f_c, sigma):
    """
    Generate a Morlet wavelet.

    Parameters:
    t (np.array): Time vector.
    f_c (float): Central frequency.
    sigma (float): Standard deviation of the Gaussian envelope.

    Returns:
    np.array: Morlet wavelet.
    """
    
    norm_factor = (1 / np.pi**0.25)
    return  norm_factor * np.exp(1j * 2 * np.pi * f_c * t) * np.exp(-t**2 / (2 * sigma**2))
```

```python
# Show the morlet wavelet
t = np.linspace(-1, 1, 1000)
wavelet1 = np.real(morlet_wavelet(t, f_c=1, sigma=0.1))
wavelet2 = np.real(morlet_wavelet(t, f_c=2.5, sigma=0.5))
wavelet3 = np.real(morlet_wavelet(t, f_c=5, sigma=0.9))

plt.figure(figsize=(10, 6))
plt.plot(t, wavelet1)
plt.plot(t, wavelet2)
plt.plot(t, wavelet3)
plt.legend(['f_c=0.5', 'f_c=1', 'f_c=1.5'])
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.show()
```

```python
def generate_wavelets(n_wavelets: int, f_c_range: Tuple[float, float], sigma_range: Tuple[float, float]) -> np.ndarray:
    """
    Generate a set of Morlet wavelets.

    Parameters:
    n_wavelets (int): Number of wavelets to generate.
    f_c_range (Tuple[float, float]): Range of central frequencies.
    sigma_range (Tuple[float, float]): Range of standard deviations.
    
    Returns:
    np.ndarray: Array of wavelets.
    """
    
    wavelets = []
    for _ in range(n_wavelets):
        f_c = np.random.uniform(*f_c_range)
        sigma = np.random.uniform(*sigma_range)
        wavelet = morlet_wavelet(t, f_c, sigma)
        wavelets.append(np.real(wavelet))
    return wavelets
```
